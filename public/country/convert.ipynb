```python
import pandas as pd
```

```python
# cnt_code.csv 파일을 읽어와서 df로 변경
df = pd.read_csv("cnt_code.csv", encoding="utf-16")

df
```

```python
cols = ['국가명(영문)', '국가명(국문)', 'ISO(2자리)']
df_split = df.iloc[:, 0].str.split('\t', expand=True)  # 탭 기준으로 펼치기
df_split.columns = cols
mapping = dict(zip(
    df_split['ISO(2자리)'],
    df_split['국가명(국문)'] + "(" + df_split['국가명(영문)'] + ")"
))
```

```python
mapping
```

```python
# 딕셔너리를 json 파일로 저장
# 한글 인코딩 문제 해결
import json
with open('cnt_mapping.json', 'w', encoding='utf-8') as f:
    json.dump(mapping, f, ensure_ascii=False)
    
```
