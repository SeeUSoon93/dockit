```python
import pandas as pd
```

```python
# dogn_code.xlsx 파일을 읽어오기
df = pd.read_excel("dogn_code.xlsx")

df
```

```python
# NaN이 있는 행 제거
df = df.dropna()

# 중복 행 제거
df = df.drop_duplicates()

```

```python
df
```

```python
# 시도명, 시군구명, 읍면동명 컬럼 합치기
# 예) 서울특별시 종로구 청운효자동

df['full_dong_name'] = df['시도명'] + ' ' + df['시군구명'] + ' ' + df['읍면동명']

# 시도명, 시군구명, 읍면동명 컬럼 제거
df = df.drop(columns=['시도명', '시군구명', '읍면동명'])

# full_dong_name 컬럼 이름 변경
df = df.rename(columns={'full_dong_name': 'full_dong_name'})

```

```python
df
```

```python
# 행정동코드 컬럼이 key, full_dong_name 컬럼이 value인 딕셔너리 생성
dong_dict = dict(zip(df['행정동코드'], df['full_dong_name']))
```

```python
dong_dict
```

```python
# 딕셔너리를 json 파일로 저장
# 한글 인코딩 문제 해결
import json
with open('dong_dict.json', 'w', encoding='utf-8') as f:
    json.dump(dong_dict, f, ensure_ascii=False)
    
```
